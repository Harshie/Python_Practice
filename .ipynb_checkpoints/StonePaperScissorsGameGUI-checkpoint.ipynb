```python
import PySimpleGUI as sg 
import random

# Game options
options = ["rock","paper","scissors"]   

# Function for comparing the user and cumputer's options.
def compare(userInput,systemInput):
    if userInput == systemInput:
        window['-OUTPUT-'].update("It's a tie. Please try again!")
    
    elif (userInput == "paper" and systemInput == "rock") or (userInput == "rock" and systemInput == "scissors") or (userInput == "scissors" and systemInput == "paper"):
        window['-OUTPUT-'].update("And YOU WON!")
    
    elif (systemInput == "paper" and userInput == "rock") or (systemInput == "rock" and userInput == "scissors") or (systemInput == "scissors" and userInput == "paper"):
        window['-OUTPUT-'].update("And COMPUTER WON!")


# GUI Theme
sg.theme('GreenMono') 

# GUI Layout
layout = [[sg.Text('Please select your choice',font = ("Helvetica", 15),justification='center')],
          [sg.T("")],
          [sg.Button('ROCK', size = (7, 2), font = ("Helvetica", 15)), sg.Button('PAPER',size=(8, 2),font=("Helvetica", 15)), sg.Button('SCISSORS',size=(10, 2),font=("Helvetica", 15))],
          [sg.T("")],
          [sg.T("")],
          [sg.Text(size=(50,1), key='-USER_OPTION-',font = ("Helvetica", 15),justification='center')],
          [sg.Text(size=(50,1), key='-COMPUTER_OPTION-',font = ("Helvetica", 15),justification='center')],
          [sg.Text(size=(30,1), key='-OUTPUT-',font = ("Helvetica", 15),justification='center')],  
          [sg.T("")],
          [sg.T("")],
          [sg.Button('Try Again', size = (8, 2), font = ("Helvetica", 15)), sg.Button('Exit', size = (7, 2), font = ("Helvetica", 15))]]

# Create the window
window = sg.Window('Roshambo game', layout, size=(500,600),element_justification='c') 

# Event loop
while True: 
    event, values = window.read() 
    print(event, values) 
      
    if event in  (None, 'Exit'): 
        break
    
    # Fetching user response
    if event == 'ROCK':
        userInput = 'rock'
        print("ROCK")
    if event == 'PAPER':
        userInput = 'paper'
    if event == 'SCISSORS':
        userInput = 'scissors'
    
            
    # Fetching random choice
    systemInput = random.choice(options).lower()

    #Displaying values to screen
    window['-USER_OPTION-'].update("Your move was: {}".format(userInput.upper()))
    window['-COMPUTER_OPTION-'].update("Computer's move was: {}".format(systemInput.upper()))

    # Calling function for comparison
    compare(userInput,systemInput)
    
    # Clearing the output for playing again
    if event == 'Try Again':
        window.FindElement('-USER_OPTION-').Update('')
        window.FindElement('-COMPUTER_OPTION-').Update('')
        window.FindElement('-OUTPUT-').Update('')

# Close the window        
window.close() 
```
